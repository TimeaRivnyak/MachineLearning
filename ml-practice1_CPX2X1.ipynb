```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

data=pd.read_csv('../input/heart-diseasecsv/heart_disease.csv')
label=data.values[:,0] #Y
data=data.values[:,1:] #X
```

```python
class LinearRegression():
    def __init__(self):
        self.w_hat = None
        self.bias = None
        
    def fit(self,data,labels):
        extended_data = np.column_stack((data, np.ones(data.shape[0])))
        w = np.linalg.inv(extended_data.T @ extended_data) @ extended_data.T @ labels
        self.w_hat = w[:-1]
        self.bias = w[-1]
        
    def fit_without_bias(self,data,labels):
        self.w_hat = np.linalg.inv(data.T @ data) @ data.T @ labels

    def predict(self,data):
        return (data @ self.w_hat + self.bias)
    
    def predict_without_bias(self,data):
        return (data @ self.w_hat)

    def mean_squared_error(self,true,prediction):
        return ((true - prediction)**2).mean()

    def mean_absolute_percentage_error(self,true,prediction):
        return np.absolute((true - prediction) / prediction).mean()*100

    def mean_absolute_error(self,true,prediction):
        return np.absolute(true - prediction).mean()

    
```

```python
linreg=LinearRegression()
linreg.fit(data,label)
prediction=linreg.predict(data)

print(linreg.mean_squared_error(label,prediction))
print(linreg.mean_absolute_error(label,prediction))
print(linreg.mean_absolute_percentage_error(label,prediction))
```

```python
linreg=LinearRegression()
linreg.fit_without_bias(data,label)
prediction_without_bias=linreg.predict_without_bias(data)

print(linreg.mean_squared_error(label,prediction_without_bias))
print(linreg.mean_absolute_error(label,prediction_without_bias))
print(linreg.mean_absolute_percentage_error(label,prediction_without_bias))
```
