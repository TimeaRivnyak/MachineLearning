```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt

data=pd.read_csv('../input/heart-diseasecsv/heart_disease.csv')
labels=data.values[:,-1]
labels[labels>1]=1
labels[labels==0]=-1
labels=labels.astype(int)

data=data.values[:,:-1]
```

```python
def visualize(data,labels,w):
    #pass
    #TODO
    data_v=data[:,[3,7]]
    w_v = w[[3,7,-1]]
    def separator_points(w_v):
        l = -w_v[1] / (w_v[2]**2 + w_v[0]**2)
        x0 = np.array([l * w_v[2], l * w_v[0]])
        v = np.array([-w_v[0], w_v[2]])
        return np.stack([x0 + v, x0 - v])
    fig = plt.figure()
    plt.plot(data_v[:, 0][labels == 1], data_v[:, 1][labels == 1], 'x', label='+1')
    plt.plot(data_v[:, 0][labels == -1], data_v[:, 1][labels == -1], 'o', label='-1')
    x_min = np.amin(data_v[:, 0])
    x_max = np.amax(data_v[:, 0])
    plt.xlim([x_min-10,x_max+10])
    y_min = np.amin(data_v[:, 1])
    y_max = np.amax(data_v[:, 1])
    plt.ylim([y_min-10,y_max+10])
    s = separator_points(w_v)
    plt.plot(s[:, 0], s[:, 1], '-', label='separator')
    plt.legend()
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.title('Perceptron Learning Algorithm')
    plt.show()
```

```python
class Perceptron():
    def __init__(self):
        self.w_hat = None
        self.bias = None
        
    def fit(self,data,labels,max_iterations=500):
        #TODO
        extended_data = np.column_stack((data, np.ones(data.shape[0])))
        w = np.zeros(extended_data.shape[1])
        count = 0
        done = False
        fig, ax = None, None
        while (not done) and (count < max_iterations):
            done = True
            count += 1
            for x, y in zip(extended_data, labels):
                if np.sign(x @ w) != y:
                    done = False
                    w = w + y*x
            if count%100 == 0:
                visualize(extended_data, labels, w)
        self.w_hat = w[:-1]
        self.bias = w[-1]
        #pass

    def predict(self,data):
        return np.sign(data @ self.w_hat + self.bias)
        #pass
    
    def accuracy(self,true,prediction):
        #TODO
        return (prediction == true).mean()
        #pass
    
    def sensitivity(self,true,prediction):
        #TODO
        TP = 0
        FN = 0
        for t, p in zip(true, prediction):
            if t == 1 and p == 1:
                TP += 1
            elif t == 1 and p == -1:
                FN += 1
        return TP/(TP+FN)
        #pass
```

```python
pla=Perceptron()
pla.fit(data,labels)
prediction=pla.predict(data)
print(pla.accuracy(labels,prediction))
print(pla.sensitivity(labels,prediction))
```
